```python
from codecarbon import EmissionsTracker
from pathlib import Path

# Setting the project root manually
PROJECT_ROOT = Path(r"C:\Users\ZEUS\Documents\carbon_aware_marine_ai")

# Creating the emissions output folder inside the project
emissions_dir = PROJECT_ROOT / "outputs" / "emissions"
emissions_dir.mkdir(parents=True, exist_ok=True)

print("Emissions will be saved to:")
print(emissions_dir)

tracker = EmissionsTracker(
    project_name="marine_carbon_ai_environment_test",
    output_dir=str(emissions_dir),
    output_file="test_emissions.csv",
    log_level="error"
)

tracker.start()

total = 0
for i in range(5_000_000):
    total += i ** 2

emissions = tracker.stop()

print("Computation finished.")
print("Test result:", total)
print("Estimated emissions kg CO2eq:", emissions)

# Confirming whether the file exists
csv_path = emissions_dir / "test_emissions.csv"

print("\nCSV file path:")
print(csv_path)

print("\nCSV exists:", csv_path.exists())
```
